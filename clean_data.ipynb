```python
import pandas as pd
import demoji
import re
from pythainlp import word_tokenize
import matplotlib.pyplot as plt
```

```python
#รับ data set เข้ามา

get_data_from_excel = pd.read_excel(r'C:\Users\your username\Desktop\shopee\shopee_scraping_dropDup.xlsx')  
```

```python
get_data_from_excel
```

```python
#ลบ emoji และ สัญลักษ์ต่าง ๆ ออกจาก column Products

demoji.download_codes()

def remove_emojis(text):
    return demoji.replace_with_desc(text, '')

def remove_symbol(text):
    regex_pattern = r'[0-9]+|[^ก-๙a-zA-Z\s]'
    removed_symbol = re.sub(regex_pattern, '', text)
    return removed_symbol

get_data_from_excel['Products'] = get_data_from_excel['Products'].apply(remove_emojis)
get_data_from_excel['Products'] = get_data_from_excel['Products'].apply(remove_symbol)

print(get_data_from_excel)
```

```python
#ลบ stop word หรือคำที่เป็นแค่ส่วนเสริมของประโยค คำที่เราไม่สนใจให้มีใน products เพื่อนำไปใช้งาน
#split หรือ แยกคำให้ออกเป็นส่วน ๆ คล้ายการเรียงประโยคแบบภาษาอังกฤษ

stop_words = ['มี', 'ที่', 'สำหรับ', 'แบบ', 'ได้', 'ดี', 'และ', 'พร้อม', 'x', 'X']

def split_text(text):
    word_split = "".join(i for i in text if i not in ('"', "ๆ", "ฯ"))
    word_split = word_tokenize(word_split)
    word_split = " ".join(i for i in word_split)
    word_split = " ".join(i for i in word_split.split()
                         if i not in stop_words)
    return word_split

get_data_from_excel['Products'] = get_data_from_excel['Products'].apply(split_text)
```

```python
#แยกราคาที่อยู่ใน format ฿xxx - ฿xxx แบ่ง column เป็น min price และ max price สำหรับราคาที่ไม่ได้อยู่ใน format นี้ก็ใช้ค่าเดียวกันทั้ง 2 column

def extract_minPrice(text):
    if " - " in text:
        split_value = text.split(" - ")  
        min_price = int(split_value[0][1:].replace(',', '')) #replace , ด้วยค่าว่างเปล่าเพื่อเช็คค่าที่มีราคาตั้งแต่ 1,000 ขึ้นไป
        max_price = int(split_value[1][1:].replace(',', ''))
        return min(min_price, max_price)  
    else:
        
        return int(text[1:].replace(',', ''))

def extract_maxPrice(text):
    if " - " in text:
        split_value = text.split(" - ")  
        min_price = int(split_value[0][1:].replace(',', ''))
        max_price = int(split_value[1][1:].replace(',', ''))
        return max(min_price, max_price)  
    else:
        
        return int(text[1:].replace(',', ''))


get_price = get_data_from_excel['Price']

minPrice_list = pd.DataFrame(get_price) 
maxPrice_list = pd.DataFrame(get_price)

minPrice_list['Price'] = minPrice_list['Price'].apply(extract_minPrice)
maxPrice_list['Price'] = maxPrice_list['Price'].apply(extract_maxPrice)
```

```python
#สร้าง column ขึ้นมา และ insert เข้าไปในตารางที่มีอยู่ก่อนแล้ว

get_data_from_excel['Min-price'] = minPrice_list
get_data_from_excel['Max-price'] = maxPrice_list

data = get_data_from_excel.columns.tolist()
data.insert(2 , data.pop(data.index('Min-price')))
get_data_from_excel = get_data_from_excel[data]

data = get_data_from_excel.columns.tolist()
data.insert(3 , data.pop(data.index('Max-price')))
get_data_from_excel = get_data_from_excel[data]

get_data_from_excel = get_data_from_excel[data]
get_data_from_excel
```

```python
#ลบ column price ที่ไม่ได้ใช้แล้วออก

get_data_from_excel = get_data_from_excel.drop('Price', axis=1)
```
